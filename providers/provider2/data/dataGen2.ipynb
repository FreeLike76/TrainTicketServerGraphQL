# Imports

```python
import pandas as pd
import numpy as np
import datetime
```

# Config

```python
np.random.seed(1)
```

# Known data

```python
dates = np.array(["2021-11-11", "2021-11-12", "2021-11-13"])

times = np.array(["00:00", "07:00", "12:00", "18:00", "20:00"])

cities = np.array(["Lviv", "Rivne", "Kyiv", "Kharkiv", "Odessa"])

# (type 1, type 2, type3) per train
type_count = np.array([50, 50, 50])
```

```python
df_trips = pd.DataFrame(columns=
[
    "id",
    "date",
    "time",
    "from_city",
    "to_city",
    "travel_time",
    "seat_type_1_price",
    "seat_type_2_price",
    "seat_type_3_price"
    ])
df_trips["id"] = [x for x in range(1, 26)]
df_trips["date"] = dates[np.random.randint(0, dates.shape[0], 25)]
df_trips["time"] = times[np.random.randint(0, times.shape[0], 25)]
df_trips["from_city"] = cities[np.random.randint(0, cities.shape[0], 25)]
df_trips["to_city"] = cities[np.random.randint(0, cities.shape[0], 25)]

df_trips["travel_time"] = np.random.randint(3, 10, 25)

df_trips["seat_type_1_price"] = np.round(np.random.normal(100, 20, 25), 2)
df_trips["seat_type_2_price"] = np.round(np.random.normal(100, 20, 25), 2)
df_trips["seat_type_3_price"] = np.round(np.random.normal(100, 20, 25), 2)
df_trips
```

```python
df_tickets = pd.DataFrame(columns=["trip_id", "seat_type", "seat_num"])

for trip_i, trip_row in df_trips.iterrows():
    
    df_trip_tickets = pd.DataFrame(columns=["trip_id", "seat_type", "seat_num"])

    df_trip_tickets["trip_id"] = [trip_i + 1 for x in range(type_count.sum())]
    df_trip_tickets["seat_type"] = [1 for x in range(type_count[0])] + [2 for x in range(type_count[1])] + [3 for x in range(type_count[2])]
    df_trip_tickets["seat_num"] = list(range(type_count[0])) + list(range(type_count[1])) + list(range(type_count[2]))

    df_tickets = df_tickets.append(df_trip_tickets, ignore_index=True)
df_tickets["id"] = [x for x in range(1, len(df_tickets) + 1)]
df_tickets
```

## Save

```python
df_trips.to_csv("trips.csv", index=False)
```

```python
df_tickets.to_csv("tickets.csv", index=False)
```
